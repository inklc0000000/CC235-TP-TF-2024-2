```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
from google.colab import drive
```

```python
# Cargar el modelo
drive.mount('/content/drive')
model = tf.keras.models.load_model('/content/drive/MyDrive/TF_PI/modelo_densenet.keras')
print("Modelo cargado correctamente.")
```

```python

# Función para predecir la categoría de una imagen
def predecir_imagen(ruta_imagen, class_names):
    # Cargar la imagen y redimensionarla al tamaño esperado
    img = load_img(ruta_imagen, target_size=(128, 128))

    # Convertir la imagen a un array y escalar los valores
    img_array = img_to_array(img) / 255.0

    # Añadir una dimensión para que coincida con la entrada del modelo (batch_size, height, width, channels)
    img_array = np.expand_dims(img_array, axis=0)

    # Realizar la predicción
    predicciones = model.predict(img_array)

    # Obtener la categoría con mayor probabilidad
    clase_predicha = np.argmax(predicciones, axis=1)[0]
    return class_names[clase_predicha]

```

```python
# Lista de nombres de las clases
class_names = ['Cardboard', 'Food Organics', 'Glass', 'Metal', 'Miscellaneous Trash', 'Paper', 'Plastic', 'Textile Trash', 'Vegetation']
```

```python

# Probar con una imagen
ruta_imagen_prueba = '/content/drive/MyDrive/TF_PI/imagen_prueba.png'  # Cambia a la ruta de tu imagen
categoria = predecir_imagen(ruta_imagen_prueba, class_names)
print(f"La imagen pertenece a la categoría: {categoria}")
```

```python
def predecir_probabilidades(ruta_imagen, class_names):
    # Cargar la imagen y redimensionarla al tamaño esperado
    img = load_img(ruta_imagen, target_size=(128, 128))

    # Convertir la imagen a un array y escalar los valores
    img_array = img_to_array(img) / 255.0

    # Añadir una dimensión para que coincida con la entrada del modelo (batch_size, height, width, channels)
    img_array = np.expand_dims(img_array, axis=0)

    # Realizar la predicción
    predicciones = model.predict(img_array)

    # Convertir a porcentajes
    probabilidades = predicciones[0] * 100

    # Mostrar los resultados para todas las clases
    for i, prob in enumerate(probabilidades):
        print(f"Clase: {class_names[i]}, Probabilidad: {prob:.2f}%")

    # Retornar las probabilidades en forma de diccionario (opcional)
    return dict(zip(class_names, probabilidades))
```

```python
resultados = predecir_probabilidades(ruta_imagen_prueba, class_names)
```
